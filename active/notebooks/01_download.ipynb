```python
import os
os.chdir(os.path.dirname(os.getcwd()) if os.path.basename(os.getcwd()) == "notebooks" else os.getcwd())

config_path = "configs/central-kalimantan.yaml"

# ── Resume control ────────────────────────────────────────────────────────────
# resume = False  → create a fresh timestamped run folder (first run)
# resume = True   → reload the most recent matching run folder (after kernel
#                   restart or error mid-notebook; no data is re-downloaded
#                   unless use_cache below also forces it)
resume = True

# ── Cache control (ignored when data already exists in the run folder) ────────
use_cache = {"forest": True, "variables": True, "mill": True}

from palmdef_risk.io.run import create_or_resume_run
ctx = create_or_resume_run(config_path, resume=resume)
```

```python
from palmdef_risk.cache import CacheManager
from palmdef_risk.io.helpers import aoi_bbox_4326

cm = CacheManager(ctx.config.cache_dir)
bbox = aoi_bbox_4326(ctx.config.aoi_source)
t2, t3 = ctx.config.forest_years[1], ctx.config.forest_years[2]
forest_key = cm.forest_key(bbox, ctx.config.aoi_buffer,
                           ctx.config.forest_source,
                           ctx.config.forest_years, ctx.config.forest_perc)
vars_key = cm.variables_key(bbox, ctx.config.aoi_buffer,
                            ctx.config.use_ghsl_towns,
                            ctx.config.ghsl_years, ctx.config.osm_timeout,
                            ctx.config.river_source)
report = cm.status_report(t2, t3, bbox, forest_key, vars_key)
for dataset, status in report.items():
    print(f"{dataset}: {status}")
```
